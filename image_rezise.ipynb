```python
import glob
import cv2
import os
import imutils
import matplotlib.pyplot as plt

# inputFolder = "/home/nuc-obs-06/projects/Aravind/Test_Samples_Practice/us_samples/DataSets/original_dataset"
# outputFolder = "/home/nuc-obs-06/projects/Aravind/Test_Samples_Practice/us_samples/DataSets/resized_dataset"

#trevor
inputFolder = "/home/nuc-obs-06/projects/Aravind/Workspace/concepts_handson/data/input"
outputFolder = "/home/nuc-obs-06/projects/Aravind/Workspace/concepts_handson/data/output"
#dl resize
# dim = (600,400)
#selfi resize
dim = (640,480)
width = 1024
height = 1024
inter = cv2.INTER_AREA
for imgName in os.listdir(inputFolder):
    image = cv2.imread(inputFolder+"/"+imgName)
    resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
    if not os.path.exists(outputFolder):
        os.mkdir(outputFolder)

    %matplotlib inline
    plt.imshow(cv2.cvtColor(resized,cv2.COLOR_RGB2BGR))
    plt.show()


    cv2.imwrite(outputFolder+"/"+imgName,resized)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
