```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
import pytesseract

img = cv2.imread('/home/nuc-obs-06/projects/Aravind/Workspace/dl1.jpg',0)
original_image = img.copy()
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
# plt.imshow(img)
# plt.show()
```

```python
image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
ret,thresh_binary = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh_binary)
```

```python
ret,thresh_binary = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh_binary)
```

```python
ret,thresh_binary = cv2.threshold(image,127,255,cv2.THRESH_TRUNC)
plt.imshow(thresh_binary)
```

```python
ret,thresh_binary = cv2.threshold(image,127,255,cv2.THRESH_TOZERO)
plt.imshow(thresh_binary)
```

```python
ret,thresh_binary = cv2.threshold(image,127,255,cv2.THRESH_TOZERO_INV)
plt.imshow(thresh_binary)
```

```python
im_gray = cv2.imread('/home/nuc-obs-06/projects/Aravind/Workspace/dl1.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
(thresh, im_bw) = cv2.threshold(im_gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
cv2.imshow('image',im_bw)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(im_bw)
```

```python
crop_img = im_bw[120:430, 240:650] # Crop from {x, y, w, h } => {0, 0, 300, 400}
plt.imshow(crop_img)
# cv2.imshow('fsdf',crop_img)
# cv2.waitKey(0)
```

```python
# output = pytesseract.image_to_string(Image.fromarray(crop_img))
output = pytesseract.image_to_string(Image.fromarray(crop_img))
print(output)
```
